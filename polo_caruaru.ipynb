#Criando um DF de exportações

```python
# Filtrar apenas dados de exportação (ignorando vendas nacionais, se você preencheu como 'Nacional')
df_exportacoes = df_polo_caruaru[df_polo_caruaru['País de Exportação'] != 'Nacional']

# Agrupar por país e gênero, contando o número de exportações
export_por_genero = df_exportacoes.groupby(['País de Exportação', 'Gênero do Consumidor']).size().unstack(fill_value=0)

# Calcular proporção de feminino em relação ao total de exportações (feminino + masculino)
export_por_genero['Proporção Feminino'] = export_por_genero['feminino'] / (export_por_genero['feminino'] + export_por_genero['masculino'])

# Ordenar pela maior proporção
export_por_genero = export_por_genero.sort_values(by='Proporção Feminino', ascending=False)

# Exibir o país com maior proporção
print("País com maior proporção de exportações femininas:")
print(export_por_genero[['Proporção Feminino']].head(1))
```

Ou seja, os EUA tem uma taxa de 63% de exportações femininas, mesmo nao sendo o pais que MAIS exporta, mas é o país que mais exporta roupas feninas em relaçao aos outros

## Analisando os piores produtos do catalogo


```python


# Agrupar os dados por produto e calcular as métricas solicitadas
produtos_desempenho = df_polo_caruaru.groupby('Produto').agg(
    volume_vendas=('Quantidade Vendida', 'sum'),
    receita_total=('Receita Total (R$)', 'sum'),
    tempo_producao_medio=('Tempo de Produção por Peça (min)', 'mean')
).reset_index()

# Adicionar uma coluna para comparar o desempenho
produtos_desempenho['Desempenho'] = (
    produtos_desempenho['volume_vendas'] / produtos_desempenho['volume_vendas'].max() +
    produtos_desempenho['receita_total'] / produtos_desempenho['receita_total'].max() +
    produtos_desempenho['tempo_producao_medio'] / produtos_desempenho['tempo_producao_medio'].max()
)

# Ordenar os produtos por desempenho para ver qual tem o pior desempenho
produtos_desempenho_sorted = produtos_desempenho.sort_values(by='Desempenho', ascending=True)

# Exibir os resultados
print("Tabela de Desempenho dos Produtos:")
print(produtos_desempenho_sorted[['Produto', 'volume_vendas', 'receita_total', 'tempo_producao_medio', 'Desempenho']])

# Produto com pior desempenho (menor pontuação de desempenho)
produto_pior_desempenho = produtos_desempenho_sorted.iloc[0]
print("\nProduto com o pior desempenho combinado:")
print(produto_pior_desempenho[['Produto', 'volume_vendas', 'receita_total', 'tempo_producao_medio', 'Desempenho']])
```

```python
# Cálculo do novo índice de desempenho
df_polo_caruaru['Desempenho Refinado'] = (df_polo_caruaru['Tempo de Produção por Peça (min)']) / (df_polo_caruaru['Quantidade Vendida'] * df_polo_caruaru['Receita Total (R$)'])

# Exibindo a tabela com o novo desempenho
tabela_desempenho_refinado = df_polo_caruaru[['Produto', 'Quantidade Vendida', 'Receita Total (R$)', 'Tempo de Produção por Peça (min)', 'Desempenho Refinado']]
print(tabela_desempenho_refinado)

# Ordenando pela menor performance (onde menor é melhor)
tabela_desempenho_refinado = tabela_desempenho_refinado.sort_values(by='Desempenho Refinado', ascending=True)

# Exibindo a empresa com o melhor desempenho
print("\nProduto com o melhor desempenho refinado:")
print(tabela_desempenho_refinado.head(1))
```

```python
import matplotlib.pyplot as plt

# Plotando o gráfico de desempenho refinado
plt.figure(figsize=(10, 6))
plt.bar(tabela_desempenho_refinado['Produto'], tabela_desempenho_refinado['Desempenho Refinado'], color='skyblue')
plt.title('Desempenho Refinado dos Produtos')
plt.xlabel('Produto')
plt.ylabel('Desempenho Refinado')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Vamos supor que você tenha uma coluna 'Data da Venda' e uma coluna 'Produto' no seu DataFrame
# Primeiro, vamos garantir que a coluna 'Data da Venda' seja do tipo datetime

df_polo_caruaru['Data da Venda'] = pd.to_datetime(df_polo_caruaru['Data da Venda'])

# Filtrar os dados para o mês de dezembro de cada ano (2022, 2023, 2024)
df_dezembro = df_polo_caruaru[df_polo_caruaru['Data da Venda'].dt.month == 12]

# Agrupar por Produto e ano, e somar as vendas (assumindo que a coluna 'Quantidade Vendida' esteja correta)
df_dezembro['Ano'] = df_dezembro['Data da Venda'].dt.year
vendas_dezembro = df_dezembro.groupby(['Produto', 'Ano'])['Quantidade Vendida'].sum().unstack()

# Exibindo as vendas por produto e ano
print(vendas_dezembro)

# Agora, vamos analisar o produto com o maior padrão de vendas consistente em dezembro
# O produto com a maior soma de vendas em dezembro será o escolhido
produto_mais_vendido = vendas_dezembro.sum(axis=1).idxmax()
print(f"Produto com maior consistência de vendas em dezembro: {produto_mais_vendido}")
```

#Padrões Sazonais por tipo de Produto

```python
# Garantindo que a coluna de data esteja no formato datetime
df_polo_caruaru['Data da Venda'] = pd.to_datetime(df_polo_caruaru['Data da Venda'], dayfirst=True)

# Criando colunas auxiliares de mês e ano
df_polo_caruaru['Mês'] = df_polo_caruaru['Data da Venda'].dt.month
df_polo_caruaru['Ano'] = df_polo_caruaru['Data da Venda'].dt.year

# Filtrando apenas as vendas feitas no mês de dezembro
df_dezembro = df_polo_caruaru[df_polo_caruaru['Mês'] == 12]

# Agrupando por Produto e Ano para contar quantas unidades foram vendidas em dezembro
vendas_dezembro = df_dezembro.groupby(['Produto', 'Ano'])['Quantidade Vendida'].sum().unstack()

# Exibindo a tabela
print("📅 Vendas de dezembro por produto:")
display(vendas_dezembro)

# Analisando qual produto aparece com vendas consistentes em todos os anos
vendas_dezembro['Total Anos com Vendas'] = vendas_dezembro.notna().sum(axis=1)
produto_mais_consistente = vendas_dezembro['Total Anos com Vendas'].idxmax()

print(f"\n✅ Produto com maior consistência de vendas em dezembro: **{produto_mais_consistente}**")
```

#Tempo Ocioso de Máquina

```python
# Agrupar por empresa e somar o tempo de produção e tempo ocioso
df_ociosidade = df_polo_caruaru.groupby('Empresa')[['Tempo de Produção por Peça (min)', 'Tempo Ocioso das Máquinas (min)']].sum()

# Calcular o tempo total
df_ociosidade['Tempo Total (min)'] = df_ociosidade['Tempo de Produção por Peça (min)'] + df_ociosidade['Tempo Ocioso das Máquinas (min)']

# Calcular a proporção de tempo ocioso
df_ociosidade['Proporção Ociosidade'] = df_ociosidade['Tempo Ocioso das Máquinas (min)'] / df_ociosidade['Tempo Total (min)']

# Ordenar pela maior ociosidade
df_ociosidade_sorted = df_ociosidade.sort_values(by='Proporção Ociosidade', ascending=False)

# Exibir resultado
print("📊 Proporção de Ociosidade por Empresa:\n")
print(df_ociosidade_sorted[['Proporção Ociosidade']])

# Mostrar a empresa com maior ociosidade
empresa_maior_ociosidade = df_ociosidade_sorted[['Proporção Ociosidade']].idxmax()
print("\n🏭 Empresa com maior ociosidade:", empresa_maior_ociosidade.values[0])
```

#Preço médio das peças vendidas

```python
# Criando a coluna do preço médio por peça
df_polo_caruaru['Preço Médio por Peça'] = df_polo_caruaru['Preço Unitário (R$)'] / df_polo_caruaru['Quantidade Vendida']

# Agrupando por empresa e calculando a média do preço médio por peça
preço_médio_por_empresa = df_polo_caruaru.groupby('Empresa')['Preço Médio por Peça'].mean()

# Exibindo o preço médio por peça de todas as empresas formatado em R$
preço_médio_por_empresa_formatado = preço_médio_por_empresa.apply(lambda x: f"R$ {x:,.2f}")

preço_médio_por_empresa_formatado
```

```python
# Calculando a receita total por venda
df_polo_caruaru['Receita por Venda'] = df_polo_caruaru['Preço Unitário (R$)'] * df_polo_caruaru['Quantidade Vendida']

# Agrupando por empresa e somando a receita total e a quantidade total vendida
preço_médio_por_empresa = df_polo_caruaru.groupby('Empresa').agg(
    Receita_Total=('Receita por Venda', 'sum'),
    Quantidade_Total=('Quantidade Vendida', 'sum')
)

# Calculando o preço médio por peça
preço_médio_por_empresa['Preço Médio por Peça'] = preço_médio_por_empresa['Receita_Total'] / preço_médio_por_empresa['Quantidade_Total']

# Formatando o preço médio por peça para exibir em R$
preço_médio_por_empresa['Preço Médio por Peça'] = preço_médio_por_empresa['Preço Médio por Peça'].apply(lambda x: f"R$ {x:,.2f}")

# Exibindo o preço médio por peça por empresa
preço_médio_por_empresa[['Preço Médio por Peça']]
```

#Calculando o menor ticket medio de cada empresa


```python
# 1. Calcula o ticket médio por empresa:
ticket_medio_por_empresa = df_polo_caruaru.groupby('Empresa').agg(
    receita_total=('Receita Total (R$)', 'sum'),
    quantidade_vendas=('Quantidade Vendida', 'sum')
)
ticket_medio_por_empresa['ticket_medio'] = ticket_medio_por_empresa['receita_total'] / ticket_medio_por_empresa['quantidade_vendas']


# 2. Calcula o faturamento médio por empresa usando o ticket médio:
faturamento_medio_por_empresa = ticket_medio_por_empresa.groupby('Empresa')['ticket_medio'].mean()

# 3. Encontra a empresa com o menor faturamento médio:
menor_faturamento_medio = faturamento_medio_por_empresa.min()
empresa_menor_faturamento = faturamento_medio_por_empresa.idxmin()

# 4. Exibe o resultado:
print(f"A empresa com o menor faturamento médio é: {empresa_menor_faturamento}, com um faturamento médio de R$ {menor_faturamento_medio:.2f}")
```
